```python
# Imports 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ash = pd.read_csv('pokemon.csv')
```

### Exploring dataset 

```python
ash.info()
```

```python
ash.columns
```

```python
ash.isnull().sum()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(ash.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```

```python
missing = ash.isnull()
corr = missing.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation of Missingness')
plt.show()
```
